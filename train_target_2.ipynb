```python
# Вычисляем процент пропущенных значений для каждого признака
missing_percent = train_data_target2.isnull().sum() * 100 / len(train_data_target2)

# Определяем признаки, где процент пропущенных значений больше 70
features_to_drop = missing_percent[missing_percent > 50].index

# Удаляем эти признаки из датафрейма
train_data_target2 = train_data_target2.drop(features_to_drop, axis=1)
```

```python
import numpy as np

# Заполнение пропущенных значений медианой для числовых признаков
for col in train_data_target2.select_dtypes(include=np.number).columns:
    train_data_target2[col].fillna(train_data_target2[col].median(), inplace=True)
```

```python
X = train_data_target2.drop('target_2', axis=1)
y = train_data_target2['target_2']
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# Для кросс-валидации нам нужно убедиться, что в каждом фолде будет одинаковое соотношение классов
# Для этого мы используем imblearn.pipeline вместо обычного sklearn.pipeline
over = SMOTE(sampling_strategy=0.1)  # Over-sampling
under = RandomUnderSampler(sampling_strategy=0.5)  # Under-sampling

# Объединяем в один pipeline
pipeline = Pipeline(steps=[('o', over), ('u', under)])

# Применяем к нашим данным
X_res, y_res = pipeline.fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X_res.с_date_year.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from xgboost import XGBClassifier


# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Обучение модели
xgb_final = XGBClassifier(use_label_encoder=False, eval_metric="logloss")
xgb_final.fit(X_train, y_train)

# Предсказание на тестовой выборке
y_pred = xgb_final.predict(X_test)

# Вывод метрик качества
print("Матрица ошибок:")
print(confusion_matrix(y_test, y_pred))
print("\nОтчет о классификации:")
print(classification_report(y_test, y_pred))
print("Точность: ", accuracy_score(y_test, y_pred))
print("ROC AUC: ", roc_auc_score(y_test, y_pred))
```
