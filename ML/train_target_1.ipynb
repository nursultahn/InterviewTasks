```python
# Скопируем оригинальный DataFrame
# train_data_target1 = train.copy()
train_data_target1 = pd.read_csv('/content/drive/MyDrive/FreedomTest/train.csv')
# Удалим столбец target_2
train_data_target1 = train_data_target1.drop(columns='target_2')
```

```python
train_data_target1.info()
```

```python
categorical_columns = train_data_target1.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
train_data_target1['с_date'] = pd.to_datetime(train_data_target1[' с_date'])
train_data_target1['с_date_datetime'] = pd.to_datetime(train_data_target1[' с_date'])
train_data_target1['с_date_year'] = train_data_target1['с_date_datetime'].dt.year
train_data_target1['с_date_month'] = train_data_target1['с_date_datetime'].dt.month
train_data_target1['с_date_day'] = train_data_target1['с_date_datetime'].dt.day
train_data_target1['с_date_dayofweek'] = train_data_target1['с_date_datetime'].dt.dayofweek
train_data_target1.drop([' с_date'], axis=1, inplace=True)
train_data_target1.drop(['с_date_datetime'], axis=1, inplace=True)
train_data_target1.drop(['Unnamed: 0'], axis=1, inplace=True)
train_data_target1.drop(['ID_NEW'], axis=1, inplace=True)
train_data_target1.drop(['с_date'], axis=1, inplace=True)
```

```python
train_data_target1.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Вычисляем процент пропущенных значений для каждого признака
missing_percent = train_data_target1.isnull().sum() * 100 / len(train_data_target1)

# Определяем признаки, где процент пропущенных значений больше 50
features_to_drop = missing_percent[missing_percent > 50].index

# Удаляем эти признаки из датафрейма
train_data_target1 = train_data_target1.drop(features_to_drop, axis=1)
```

```python
train_data_target1.head()
```

```python
import numpy as np

# Заполнение пропущенных значений медианой для числовых признаков
for col in train_data_target1.select_dtypes(include=np.number).columns:
    train_data_target1[col].fillna(train_data_target1[col].median(), inplace=True)
```

```python
X = train_data_target1.drop('target_1', axis=1)
y = train_data_target1['target_1']
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from xgboost import XGBClassifier, XGBRegressor
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, mean_squared_error

# Создаем бинарную целевую переменную
y_binary = (y != 0).astype(int)

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test, y_binary_train, y_binary_test = train_test_split(X, y, y_binary, test_size=0.2, random_state=42)

# Применяем oversampling только для меньшего класса
ros = RandomOverSampler(sampling_strategy='minority')
X_res, y_binary_res = ros.fit_resample(X_train, y_binary_train)

# Обучаем модель классификации
classifier = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
classifier.fit(X_res, y_binary_res)

# Делаем предсказания классификации
y_binary_pred = classifier.predict(X_test)

# Выводим метрики классификации
print('Accuracy:', accuracy_score(y_binary_test, y_binary_pred))
print('F1 score:', f1_score(y_binary_test, y_binary_pred))
print('ROC AUC score:', roc_auc_score(y_binary_test, y_binary_pred))

# Обучаем модель регрессии только на ненулевых значениях
X_regression_train = X_train[y_train != 0]
y_regression_train = y_train[y_train != 0]

regressor = XGBRegressor(random_state=42)
regressor.fit(X_regression_train, y_regression_train)

# Делаем предсказания регрессии только для ненулевых предсказаний классификации
X_regression_test = X_test[y_binary_pred != 0]
y_regression_test = y_test[y_binary_pred != 0]

if len(X_regression_test) > 0:
    y_regression_pred = regressor.predict(X_regression_test)

    # Выводим метрику регрессии
    print('RMSE:', np.sqrt(mean_squared_error(y_regression_test, y_regression_pred)))
else:
    print('No non-zero predictions were made.')
```
